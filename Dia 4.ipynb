```python
#Challenge ü§∫ - 15 m
#Crear una clase animales que tenga atributo de clase y de instancia.

#Atributos de clase: especie.
#Atributos de instancia: nombre & color

#Luego, crear nuevos objetos a partir de las nuevas class. Imprimir el nombre y el color de los animales en una frase.
#Ejemplo: "Me llamo {nombre} y soy un ave de color {color}"..

class Animal:
    especie = "ave"

    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color

loro = Animal("Pepe", "Verde")
guacamayo = Animal("Dorotis", "rojo")

print (f"Me llamo {loro.nombre} y soy un ave de color {loro.color}")
print (f"Me llamo {guacamayo.nombre} y soy un ave de color {guacamayo.color}")
```

```python
#Challenge ü§∫
#Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

#Observaci√≥n: Editar las clases hijas para que no se inicialicen.

# Clase madre
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

# Clases hijas que heredan de Persona
class Estudiante(Persona):
    pass

class Profesor(Persona):
    pass

# Crear objetos de las clases hijas
estudiante = Estudiante("Ana", 20)
profesor = Profesor("Luis", 45)

# Mostrar los atributos desde los objetos de las clases hijas
print("Estudiante:")
print("Nombre:", estudiante.nombre)
print("Edad:", estudiante.edad)

print("\nProfesor:")
print("Nombre:", profesor.nombre)
print("Edad:", profesor.edad)
```

```python
#Challenge 1 ü§∫
#Escribir un programa que cree un diccionario vac√≠o y lo vaya llenado con informaci√≥n sobre una persona (por ejemplo nombre, edad, sexo, tel√©fono, correo electr√≥nico, etc.) que se le pida al usuario. 

#Cada vez que se a√±ada un nuevo dato debe imprimirse el contenido del diccionario.

# Se crea un diccionario vac√≠o para guardar los datos del usuario
datosUsuario = {}

# Definimos una funci√≥n que har√° las preguntas
def preguntas():
    # Paso 1: Pedir el nombre y apellido
    nombre = input('Ingrese su nombre y apellido: ')
    datosUsuario['Nombre'] = nombre  # Guardar el nombre en el diccionario
    print(datosUsuario)  # Mostrar el contenido actual

    # Paso 2: Pedir la direcci√≥n
    direccion = input('Ingrese su direcci√≥n: ')
    datosUsuario['Direccion'] = direccion
    print(datosUsuario)

    # Paso 3: Pedir el n√∫mero de tel√©fono
    numero = input('Ingrese su n√∫mero de tel√©fono: ')
    datosUsuario['N√∫mero'] = numero
    print(datosUsuario)

    # Paso 4: Pedir la edad y validar que sea un n√∫mero
    edad = input('Ingrese su edad: ')
    if edad.isdigit():  # Validamos que sea un n√∫mero
        datosUsuario['Edad'] = int(edad)
    else:
        print("‚ùå Edad inv√°lida. Debe ser un n√∫mero.")
    print(datosUsuario)

    # Paso 5: Mostrar resumen final
    print("\nüìù Informaci√≥n ingresada:")
    for clave, valor in datosUsuario.items():
        print(f"{clave}: {valor}")

# Paso 6: Llamamos a la funci√≥n para que se ejecute
preguntas()
```

```python
#Challenge 2 ü§∫
#Crear una tarjeta personal. El usuario (mediante input) puede elegir:

#Crear datos nuevos
#Modificar datos
#Eliminar datos

# Paso 1: Diccionario base para almacenar los datos del usuario
datos_usuario = {'Nombre': '', 'Apellido': '', 'Direccion': ''}

# Paso 2: Funci√≥n para crear una nueva tarjeta con datos ingresados por el usuario
def crear_tarjeta():
    datos_usuario['Nombre'] = input('Ingrese su nombre: ')
    datos_usuario['Apellido'] = input('Ingrese su apellido: ')
    datos_usuario['Direccion'] = input('Ingrese su direcci√≥n: ')
    print("\n‚úÖ Tarjeta creada:")
    print(datos_usuario)

# Paso 3: Funci√≥n para modificar los datos ya existentes en la tarjeta
def modificar_tarjeta():
    datos_usuario['Nombre'] = input('Ingrese nuevo nombre: ')
    datos_usuario['Apellido'] = input('Ingrese nuevo apellido: ')
    datos_usuario['Direccion'] = input('Ingrese nueva direcci√≥n: ')
    print("\n‚úÖ Tarjeta modificada:")
    print(datos_usuario)

# Paso 4: Funci√≥n para eliminar los datos si el usuario confirma
def eliminar_tarjeta():
    confirmar = input('¬øDesea eliminar todos los datos? (si/no): ')
    if confirmar.lower() == 'si':
        datos_usuario.clear()  # Elimina todos los datos del diccionario
        print("‚úÖ Datos eliminados.")
    else:
        print("‚ùå Operaci√≥n cancelada.")
    print(datos_usuario)

# Paso 5: Funci√≥n para mostrar los datos actuales de la tarjeta
def mostrar_tarjeta():
    print("\nüìÑ Tarjeta actual:")
    if datos_usuario:
        for k, v in datos_usuario.items():
            print(f"{k}: {v}")
    else:
        print("No hay datos guardados.")

# Paso 6: Bucle principal para ejecutar el men√∫ de opciones
while True:
    print("\n--- MEN√ö ---")
    print("1. Crear tarjeta")
    print("2. Modificar tarjeta")
    print("3. Eliminar tarjeta")
    print("4. Mostrar tarjeta")
    print("5. Salir")

    opcion = input("Seleccione una opci√≥n (1-5): ")

    if opcion == '1':
        crear_tarjeta()
    elif opcion == '2':
        modificar_tarjeta()
    elif opcion == '3':
        eliminar_tarjeta()
    elif opcion == '4':
        mostrar_tarjeta()
    elif opcion == '5':
        print("üëã Programa finalizado.")
        break
    else:
        print("‚ùå Opci√≥n inv√°lida. Intente nuevamente.")
```

```python
#Challenge 3 ü§∫
#Crear una agenda de contactos. El usuario puede elegir:

#Mostrar los contactos
#Crear contacto
#Modificar datos del contacto

# Paso 1: Crear un diccionario vac√≠o para guardar los contactos
agenda = {}

# Paso 2: Bucle principal para mostrar el men√∫ y ejecutar opciones
while True:
    print("\n--- AGENDA DE CONTACTOS ---")
    print("1. Mostrar contactos")
    print("2. Crear contacto")
    print("3. Modificar contacto")
    print("4. Salir")

    opcion = input("Seleccione una opci√≥n (1-4): ")

    # Paso 3: Mostrar todos los contactos
    if opcion == '1':
        print("\nüìí Lista de Contactos:")
        if agenda:
            for nombre, telefono in agenda.items():
                # Mostrar los datos en formato: Nombre: [nombre] | Tel√©fono: [telefono]
                print(f"Nombre: {nombre} | Tel√©fono: {telefono}")
        else:
            print("No hay contactos guardados.")

    # Paso 4: Crear un nuevo contacto
    elif opcion == '2':
        nombre = input("Ingrese el nombre del contacto: ")
        telefono = input("Ingrese el n√∫mero de tel√©fono: ")
        agenda[nombre] = telefono  # Agrega el nuevo contacto al diccionario
        print("‚úÖ Contacto agregado correctamente.")

    # Paso 5: Modificar un contacto existente
    elif opcion == '3':
        nombre = input("Ingrese el nombre del contacto a modificar: ")
        if nombre in agenda:
            nuevo_telefono = input("Ingrese el nuevo n√∫mero: ")
            agenda[nombre] = nuevo_telefono  # Actualiza el n√∫mero
            print("‚úÖ Contacto modificado.")
        else:
            print("‚ùå Contacto no encontrado.")

    # Paso 6: Salir del programa
    elif opcion == '4':
        print("üëã Cerrando la agenda.")
        break

    # Paso 7: Manejo de opciones inv√°lidas
    else:
        print("‚ùå Opci√≥n no v√°lida. Intente otra vez.")

```
